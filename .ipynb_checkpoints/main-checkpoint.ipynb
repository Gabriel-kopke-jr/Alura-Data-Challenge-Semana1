```python
aux_list = []
aux = ''
for i in range(dados.shape[0]):
    aux = dados['Assinatura_de_mais_linhas'].iloc[i]
    if aux == 'No':
        aux_list.append(0)
        continue
    if aux == 'Yes':    
        aux_list.append(1)
        continue
    if aux =='No phone service':
        aux_list.append(2)
dados['Assinatura_de_mais_linhas_int'] = aux_list
```

```python
aux_list = []
aux = ''
for i in range(dados.shape[0]):
    aux = dados['Assinatura_de_provedor'].iloc[i]
    if aux == 'No':
        aux_list.append(0)
        continue
    if aux == 'DSL':    
        aux_list.append(1)
        continue
    if aux =='Fiber optic':
        aux_list.append(2)
dados['Assinatura_de_provedor_int'] = aux_list
```

```python
aux_list = []
aux = ''
for i in range(dados.shape[0]):
    aux = dados['Metodo_pagamento'].iloc[i]
    if aux == 'Mailed check':
        aux_list.append(0)
        continue
    if aux == 'Electronic check':    
        aux_list.append(1)
        continue
    if aux =='Credit card (automatic)':
        aux_list.append(2)
        continue
    if aux =='Bank transfer (automatic)':
        aux_list.append(3)
dados['Metodo_pagamento_int'] = aux_list
```
